# process outliers with using interquartile range

```python
# Use IQR method to handle outliers
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)]

iqr_columns_worldometer = ["TotalCases", "TotalDeaths", "TotalRecovered", "ActiveCases", 
                           "Tot Cases/1M pop", "Deaths/1M pop", "TotalTests", "Tests/1M pop"]

# Reload Worldometer data to avoid accidental deletion of data in previous cleanup
df_worldometer_data_clean = df_worldometer_data.copy()

# use IQR method to handle outliers
df_worldometer_data_clean = remove_outliers_iqr(df_worldometer_data_clean, iqr_columns_worldometer)

print(f"Worldometer Data dataset IQR processed size: {df_worldometer_data_clean.shape}")
```

```python
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 定义函数
def remove_outliers_zscore(df, columns, threshold=3):
    df_clean = df.copy()
    for col in columns:
        z_scores = stats.zscore(df_clean[col].dropna())
        abs_z_scores = abs(z_scores)
        filtered_entries = (abs_z_scores < threshold)
        df_clean = df_clean.loc[df_clean[col].dropna().index[filtered_entries]]
    return df_clean.reset_index(drop=True)

def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean.reset_index(drop=True)

numerical_cols_full_grouped = ["Confirmed", "Deaths", "Recovered", "Active", "New cases", "New deaths", "New recovered"]
numerical_cols_covid_19 = ["Confirmed", "Deaths", "Recovered", "Active"]
iqr_columns_worldometer = ["TotalCases", "TotalDeaths", "TotalRecovered", "ActiveCases",
                           "Tot Cases/1M pop", "Deaths/1M pop", "TotalTests", "Tests/1M pop"]


df_full_grouped_clean = remove_outliers_zscore(df_full_grouped, numerical_cols_full_grouped, threshold=4)
df_covid_19_clean_clean = remove_outliers_zscore(df_covid_19_clean, numerical_cols_covid_19, threshold=4)
df_worldometer_data_clean = remove_outliers_iqr(df_worldometer_data, iqr_columns_worldometer)


# Standardization and Normalization
standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()

zscore_columns_full_grouped = numerical_cols_full_grouped
zscore_columns_covid_19 = numerical_cols_covid_19
zscore_columns_worldometer = ["TotalCases", "TotalDeaths", "TotalRecovered", "ActiveCases"]
minmax_columns_worldometer = ["Tot Cases/1M pop", "Deaths/1M pop", "TotalTests", "Tests/1M pop"]

df_full_grouped_clean[zscore_columns_full_grouped] = standard_scaler.fit_transform(df_full_grouped_clean[zscore_columns_full_grouped])
df_covid_19_clean_clean[zscore_columns_covid_19] = standard_scaler.fit_transform(df_covid_19_clean_clean[zscore_columns_covid_19])
df_worldometer_data_clean[zscore_columns_worldometer] = standard_scaler.fit_transform(df_worldometer_data_clean[zscore_columns_worldometer])
df_worldometer_data_clean[minmax_columns_worldometer] = min_max_scaler.fit_transform(df_worldometer_data_clean[minmax_columns_worldometer])

print("The complete grouped dataset after normalization (first 5 rows)：")
print(df_full_grouped_clean.head(), "\n")

print("Normalized COVID-19 Clean Complete dataset (first 5 rows)：")
print(df_covid_19_clean_clean.head(), "\n")

print("Standardized and normalized Worldometer dataset (first 5 rows)：")
print(df_worldometer_data_clean.head(), "\n")
```

```python
import pandas as pd

# Ensure date format is correct and sort
df_merged['Date'] = pd.to_datetime(df_merged['Date'])
df_merged = df_merged.sort_values(['Country/Region', 'Date'])

# Avoid division by zero - use correct column names
df_merged['Confirmed_full'] = df_merged['Confirmed_full'].replace(0, 0.001)
df_merged['Confirmed_clean'] = df_merged['Confirmed_clean'].replace(0, 0.001)
df_merged['Deaths_full'] = df_merged['Deaths_full'].replace(0, 0.001)
df_merged['Deaths_clean'] = df_merged['Deaths_clean'].replace(0, 0.001)

# 1. Basic recovery rate features - using values from full dataset
df_merged['recovery_rate_full'] = df_merged['Recovered_full'] / df_merged['Confirmed_full']
df_merged['fatality_rate_full'] = df_merged['Deaths_full'] / df_merged['Confirmed_full']
df_merged['active_rate_full'] = df_merged['Active_full'] / df_merged['Confirmed_full']

# Similarly, create features for the clean dataset
df_merged['recovery_rate_clean'] = df_merged['Recovered_clean'] / df_merged['Confirmed_clean']
df_merged['fatality_rate_clean'] = df_merged['Deaths_clean'] / df_merged['Confirmed_clean']
df_merged['active_rate_clean'] = df_merged['Active_clean'] / df_merged['Confirmed_clean']

# 2. Time-related features
df_merged['daily_new_recovered_full'] = df_merged.groupby('Country/Region')['Recovered_full'].diff().fillna(0)
df_merged['daily_new_recovered_clean'] = df_merged.groupby('Country/Region')['Recovered_clean'].diff().fillna(0)

# Avoid division by zero
df_merged['yesterday_recovered_full'] = df_merged.groupby('Country/Region')['Recovered_full'].shift(1).fillna(0.001)
df_merged['recovery_growth_full'] = (df_merged['Recovered_full'] - df_merged['yesterday_recovered_full']) / df_merged['yesterday_recovered_full']

df_merged['yesterday_recovered_clean'] = df_merged.groupby('Country/Region')['Recovered_clean'].shift(1).fillna(0.001)
df_merged['recovery_growth_clean'] = (df_merged['Recovered_clean'] - df_merged['yesterday_recovered_clean']) / df_merged['yesterday_recovered_clean']

# Calculate outbreak day for each country
df_merged['outbreak_day'] = df_merged.groupby('Country/Region').cumcount() + 1

# 3. Trend features
df_merged['recovery_rate_7d_avg_full'] = df_merged.groupby('Country/Region')['recovery_rate_full'].rolling(7).mean().reset_index(level=0, drop=True).fillna(0)
df_merged['recovery_rate_14d_avg_full'] = df_merged.groupby('Country/Region')['recovery_rate_full'].rolling(14).mean().reset_index(level=0, drop=True).fillna(0)
df_merged['recovery_rate_7d_avg_clean'] = df_merged.groupby('Country/Region')['recovery_rate_clean'].rolling(7).mean().reset_index(level=0, drop=True).fillna(0)
df_merged['recovery_rate_14d_avg_clean'] = df_merged.groupby('Country/Region')['recovery_rate_clean'].rolling(14).mean().reset_index(level=0, drop=True).fillna(0)

# Calculate trend slope
df_merged['recovery_rate_7d_trend_full'] = df_merged.groupby('Country/Region')['recovery_rate_full'].diff(7).fillna(0)
df_merged['recovery_rate_7d_trend_clean'] = df_merged.groupby('Country/Region')['recovery_rate_clean'].diff(7).fillna(0)

# Recovery acceleration
df_merged['recovery_acceleration_full'] = df_merged.groupby('Country/Region')['recovery_rate_7d_trend_full'].diff().fillna(0)
df_merged['recovery_acceleration_clean'] = df_merged.groupby('Country/Region')['recovery_rate_7d_trend_clean'].diff().fillna(0)

# 4. Ratio features
df_merged['recovery_death_ratio_full'] = df_merged['Recovered_full'] / df_merged['Deaths_full']
df_merged['healing_proportion_full'] = df_merged['daily_new_recovered_full'] / df_merged['Active_full'].replace(0, 0.001)

df_merged['recovery_death_ratio_clean'] = df_merged['Recovered_clean'] / df_merged['Deaths_clean']
df_merged['healing_proportion_clean'] = df_merged['daily_new_recovered_clean'] / df_merged['Active_clean'].replace(0, 0.001)

# New feature: Difference between datasets (potentially indicating data quality or reporting method differences)
df_merged['recovery_rate_diff'] = df_merged['recovery_rate_full'] - df_merged['recovery_rate_clean']
df_merged['fatality_rate_diff'] = df_merged['fatality_rate_full'] - df_merged['fatality_rate_clean']

# Print the first 5 rows to inspect the results
print(df_merged.head())
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# 1. Label encoding for 'Country/Region'
label_encoder = LabelEncoder()

# Encode the 'Country/Region'
df_merged['country_encoded'] = label_encoder.fit_transform(df_merged['Country/Region'])

# 2. If the dataset contains 'WHO Region' or 'Continent', apply one-hot encoding
# First, check if these columns exist
categorical_columns = []
if 'WHO Region' in df_merged.columns:
    categorical_columns.append('WHO Region')
if 'Continent' in df_merged.columns:
    categorical_columns.append('Continent')

# If categorical columns exist, apply one-hot encoding
if categorical_columns:
    from sklearn.preprocessing import OneHotEncoder
    import numpy as np
    
    # Create the one-hot encoder
    one_hot_encoder = OneHotEncoder(sparse=False, drop='first')
    
    # Apply encoding to categorical columns
    for col in categorical_columns:
        # Handle missing values
        df_merged[col] = df_merged[col].fillna('Unknown')
        
        # Apply one-hot encoding
        encoded = one_hot_encoder.fit_transform(df_merged[[col]])
        
        # Create column names for the encoded values
        encoded_cols = [f"{col}_{cat}" for cat in one_hot_encoder.categories_[0][1:]]
        
        # Add the encoded values to the dataframe
        encoded_df = pd.DataFrame(encoded, columns=encoded_cols, index=df_merged.index)
        df_merged = pd.concat([df_merged, encoded_df], axis=1)

# 3. Categorize countries into high, medium, and low groups based on confirmed cases
df_merged['country_case_group'] = pd.qcut(
    df_merged.groupby('Country/Region')['Confirmed_full'].max(), 
    q=3, 
    labels=['Low', 'Medium', 'High']
).reset_index()['Confirmed_full']

# Apply label encoding to this new feature
df_merged['country_case_group_encoded'] = label_encoder.fit_transform(df_merged['country_case_group'])

```
